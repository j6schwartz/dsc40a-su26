### Simulating the domino problem

```python
import numpy as np
```

```python
tiles = [
    f'{i}{j}' for i in range(7) for j in range(7)
]
```

```python
unique_tiles = []
for tile in tiles:
    if tile not in unique_tiles and tile[::-1] not in unique_tiles:
        unique_tiles.append(tile)
# unique_tiles += [tile[::-1] for tile in unique_tiles]
```

```python
n = 100_000

at_least_one_6 = 0
one_side_6 = 0
double_6 = 0
is_double = 0

for _ in range(n):
    tile = np.random.choice(unique_tiles)
    
    if tile in [str(i) * 2 for i in range(7)]:
        is_double += 1
    
    if '6' in tile:
        at_least_one_6 += 1
    
    # Pick a random side to look at.
    side = 1 if np.random.random() > 0.5 else 0
    if tile[side] == '6':
        one_side_6 += 1
        
    if tile == '66':
        double_6 += 1
```

```python
is_double / n
```

```python
double_6 / at_least_one_6
```

```python
double_6 / one_side_6
```

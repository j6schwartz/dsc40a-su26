```python
def single_proj(w):

    fig = plt.figure(figsize=(6, 6), dpi=200, frameon='off')

    ax = fig.add_subplot(1, 1, 1)
    # plt.grid(color='#eee')

    plt.arrow(0, 0, w*8, w*6, color='#007aff', linewidth=4, length_includes_head=True, head_width=1/16)
    plt.annotate("$" + str(w) + r"\vec{x}$", (w*8.2, w*5.8), color='#007aff', size=20)

    plt.arrow(0, 0, 2, 6, color='orange', linewidth=4, length_includes_head=True, head_width=1/16)
    plt.annotate(r"$\vec{y}$", (1.3, 5.8), color='orange', size=20)

    plt.arrow(w*8, w*6, 2-w*8, 6-w*6, color='red', linewidth=4)


    plt.xlim(0, 10)
    plt.xticks([])
    plt.ylim(0, 7.5)
    plt.yticks([])
    plt.box('off')
    # plt.savefig('../imgs/u-and-v-range.png', bbox_inches='tight')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    # plt.savefig('../imgs/project-y-on-x.png', bbox_inches='tight')
    
    plt.show()
```
